```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)
```

```python
Data_set = np.loadtxt('./dataset/ThoraricSurgery.csv', delimiter=',')
```

```python
type(Data_set)
```

```python
X = Data_set[:, 0:17]
Y = Data_set[:, 17]
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) # 첫번쩨 히든레이어의 수-노드 수(30), 첫번째 인풋의 수-17개 열에 대한 피처값(17), activation function을 태워서 출력
model.add(Dense(1, activation='sigmoid')) # 출력값(1)(생존OR사망), 인풋의 수는 전 단계에서의 출력 수(30), 최종 출력값 출력 전에 sigmoid function을 태움-이진(생존OR사망)   
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) # 이진분류: binary_crossentropy, 다중분류: categorical_crossentropy
model.fit(X, Y, epochs=100, batch_size=10)
print('\n Accuracy:%.4f' % (model.evaluate(X, Y)[1]))
```
